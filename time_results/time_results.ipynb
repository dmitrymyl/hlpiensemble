```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_context("talk")
sns.set_style("white")
```

# Length data

```python
def line_process_func(lines):
    return {i[0] + "_" + i[1]: float(i[2]) for i in [line.strip().split(":") for line in lines]}
```

```python
length_data = list()
for rna_length in range(100, 1001, 100):
    for protein_length in range(100, 1001, 100):
        with open(f"length_rna_{rna_length}_protein_{protein_length}.tsv", 'r') as infile:
            line = line_process_func(infile.readlines())
            line['RNA_length'] = rna_length
            line['protein_length'] = protein_length
            length_data.append(line)
```

```python
length_df = pd.DataFrame.from_records(length_data)
```

```python
length_df.head()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.lineplot(data=length_df, x='protein_length', y='protein_kmer')
plt.xlabel("Protein length, aa")
plt.ylabel("Protein kmer, s")
plt.subplot(1,3,2)
sns.lineplot(data=length_df, x='protein_length', y='protein_pse')
plt.xlabel("Protein length, aa")
plt.ylabel("Protein pse, s")
plt.subplot(1,3,3)
sns.lineplot(data=length_df, x='protein_length', y='protein_acc')
plt.xlabel("Protein length, aa")
plt.ylabel("Protein acc, s")
plt.tight_layout()
plt.savefig("protein_length_protein_prep.png")
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.lineplot(data=length_df, x='protein_length', y='RNA_kmer')
plt.xlabel("Protein length, aa")
plt.ylabel("RNA kmer, s")
plt.subplot(1,3,2)
sns.lineplot(data=length_df, x='protein_length', y='RNA_pse')
plt.xlabel("Protein length, aa")
plt.ylabel("RNA acc, s")
plt.subplot(1,3,3)
sns.lineplot(data=length_df, x='protein_length', y='RNA_acc')
plt.xlabel("Protein length, aa")
plt.ylabel("RNA acc, s")
plt.tight_layout()
plt.savefig("protein_length_rna_prep.png")
```

# Parallel data

```python
parallel_data = list()
for cores in range(1, 11):
    with open(f"parallel_cores_{cores}.tsv", 'r') as infile:
        line = line_process_func(infile.readlines())
        line['cores'] = cores
        parallel_data.append(line)
```

```python
parallel_df = pd.DataFrame.from_records(parallel_data)
```
